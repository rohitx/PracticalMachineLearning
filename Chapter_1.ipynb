### Choosing the right model and learning Algorithm###

Now that we have a first impression of the data, we return to the initial question: How long will our server handle the incoming web traffic? To answer this question we need to first find a model that can fit the data and then extrapolate the model into the future. 

We start with a simple straight line

```python
import scipy as sp
x = df_clean['Index_val']
y = df_clean['hits_per_hour']
fp1, residuals, rank, sv, rcond = sp.polyfit(x,y,1, full=True)
```

The polyfit() function returns the parameters of the fitted model function, fp1. And by setting full=True, we also get additional background information on the fitting process. Of this, only residuals are of interest, which is exactly the error of approximation.

```python
print "slope: {m:.4f}\nintersection: {b:.4f}".format(m = fp1[0], b=fp1[1])
```

```python
# We then use poly1d() to create a model function 
# from the model parameters:
model = sp.poly1d(fp1)

# We now plot the model with the data to see how well it fits
df_clean.plot(x = 'Index_val', y='hits_per_hour', kind='scatter', figsize=(8,6))
plt.xticks([w*7*24 for w in range(10)], ['week%i'% w for w in range(10)])
plt.autoscale(tight=True)
plt.xlabel('Weeks')
plt.ylabel('Hits/Hour')
fx = sp.linspace(0,len(x),1000)
plt.plot(fx, model(fx), linewidth=4)
plt.legend(['d=%i'%model.order], loc='upper left')
```

This is not a bad fit to the data. However, we see that the data is more of a non-linear type and so we need to fit a model that is also non-linear in nature.

```python
# Let's try a polynomial fit with orders 2, 3, 5, and 10.
f2p = sp.polyfit(x, y, 2)
f3p = sp.polyfit(x, y, 3)
f5p = sp.polyfit(x, y, 5)
f10p = sp.polyfit(x, y, 10)

model2 = sp.poly1d(f2p)
model3 = sp.poly1d(f3p)
model5 = sp.poly1d(f5p)
model10 = sp.poly1d(f10p)

# We now plot the model with the data to see how well it fits
df_clean.plot(x = 'Index_val', y='hits_per_hour', kind='scatter', figsize=(8,6))
plt.xticks([w*7*24 for w in range(10)], ['week%i'% w for w in range(10)])
plt.autoscale(tight=True)
plt.xlabel('Weeks')
plt.ylabel('Hits/Hour')
plt.plot(fx, model(fx), linewidth=3, label="d=1")
plt.plot(fx, model2(fx), linewidth=3, label="d=2")
plt.plot(fx, model3(fx), linewidth=3, label="d=3")
plt.plot(fx, model5(fx), linewidth=3, label="d=5")
plt.plot(fx, model10(fx), linewidth=3, ls = "--",label="d=10")
plt.legend(loc='upper left')
```

We find that increasing the order will result in **overfitting** while keeping a linear fit would be underfitting. Hence, looking at the plot we find that polynomial of order 2 and 3 seem to work the best. However, there is more that we can see from the data than just fitting in various polynomials. For this we need to go back a step.

We see that there is an inflection point at week 3.5. So we may need to find the data into two parts: One that is linear until Week 3.5 and then another linear point past Week 3.5. 

```python
inflection = int(3.5*7*24)
xa = x[:inflection]
ya = y[:inflection]
xb = x[inflection:]
yb = y[inflection:]

# Fit linear regression models to the two parts of the data
fa = sp.poly1d(sp.polyfit(xa, ya, 1))
fb = sp.poly1d(sp.polyfit(xb, yb, 1))


# Plot the data:
df_clean.plot(x = 'Index_val', y='hits_per_hour', kind='scatter', figsize=(10,6))
plt.xticks([w*7*24 for w in range(10)], ['week%i'% w for w in range(10)])
plt.autoscale(tight=True)
plt.xlabel('Weeks')
plt.ylabel('Hits/Hour')
plt.ylim(500, 6000)
plt.plot(fx, fa(fx), linewidth=3, label="Part_1")
plt.plot(fx, fb(fx), linewidth=3, label="Part_2")
plt.legend(loc='upper left')
```

So we see that by fitting the data with two linear regression we may seem to get a better fit. However, the computation of errors suggest that the polynomial fit with order 3 gives lower errors than linear regression. Hence, going back to the 3rd order polynomial fit we can find out when the hits will reach 100,000:

```python
df_clean.plot(x = 'Index_val', y='hits_per_hour', kind='scatter', figsize=(8,6))
plt.xticks([w*7*24 for w in range(10)], ['week%i'% w for w in range(10)])
plt.autoscale(tight=True)
plt.xlabel('Weeks')
plt.ylabel('Hits/Hour')
plt.plot(fx, model3(fx), linewidth=3, label="d=3")
plt.legend(loc='upper left')
# So the coefficients for the models are:
print f3p
```

```python
# so for x = 100,000 we have:
x = 10.3819*7*24
y = f3p[0] * x**3 + f3p[1] * x**2 + f3p[2] * x + f3p[3]
print y
```

So the answer is that it will reach 100,000 at the end of the 10th week.
